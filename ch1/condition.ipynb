`숫자가 짝수면 짝수라고 출력, 아니면 홀수라고 출력`

```python
num = 5

# if 문 : bool 자료형
if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# >, <, >=, <=, ==, != : True, False
# in, not in

a = 200

if 100 < a <=200 :
    print("a는 100과 200사이의 수이다")
else:
    print("아니다")
```

```python
# 점수가 90 이상이면 a, 80은 b, 70은 c, 나머지는 f

score = 75

if score >= 90:
    print("A")
elif 80 <= score < 90:
    print("B")
elif 70 <= score < 80: 
    print("C")
else: 
    print("F")
```

```python
a, b, c = 12, 6, 19

if a >=b and a >= c:
    print(a)
elif a <=b and b >= c:
    print(b)
else:
    print(c)
```

```python
max = a

if max <= b:
    max = b

if max <= c:
    max = c

print("가장 큰 수", max)
```

```python
min = a

if min >= b:
    min = b

if min >= c:
    min = c

print("가장 작은 수", min)
```

```python
fruits = ['딸기', '사과', '바나나', '배', '수박']

# fruits 리스트에 딸기가 포함되어 있다면 딸기가 포함됨 없다면 딸기가 없음

if '딸기' in fruits:
    print("딸기가 포함됨")
else:
    print("딸기가 없음")

```

```python
fruits.clear()

if fruits:
    print("리스트")
else:
    print("빈 리스트")

if len(fruits) == 0:
    print("빈 리스트")

if not fruits:
    print("빈 리스트")
```
