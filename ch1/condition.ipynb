```python
# 2026년이 윤년 or 평년 인지 판별

year = 2028

# 윤년의 조건
# 년도를 4로 나눈 나머지가 0 이고, 년도를 100으로 나눈ㄴ 나머지가 0이 아니거나 년도를 400으로 나눈 나머지가 0이면


if(year%4 == 0) and ((year % 100 != 0) or (year % 400 == 0)):
    print("윤년")
else:
    print("평년")

#남자 1,3 여자 2,4 판별

jumin = "001205-3234567"

idx = jumin.find('-')+1
jud = int(jumin[idx])
if(jud == 1 or jud ==3):
    print("남자")
else:
    print("여자")

# 한 시간 임금은 10700원, 8시간까지는 10700 지급, 이후부터는 1.5배 지급

hours, rate, pay = 12, 10700, 0

if hours <= 8:
    pay = rate*hours
else:
    pay = (rate*8) + (rate*1.5)*(hours-8)

print("근무시간: %d, 일당: %.0f" %(hours, pay))
print("근무시간: {0}, 일당: {1:.0f}".format(hours, pay))
print(f"근무시간: {hours}, 일당: {pay:.0f}")

```

### 반복문
- while
  
  while(조건문):
        반복 수행할 문장
        ...
        ...

- for

  for 변수 in 시퀀스:
      반복 수행할 문장
      ...
      ...


```python
# 1 ~ 10 까지 출력

i = 1
while(i<=10):
    print(i, end=" ")
    i += 1
```

```python
# 1 ~ 10 짝수만 출력

i = 1
while(i <= 10):
    if i%2 == 0:
        print(i)
    i += 1
```

```python
# 1 ~ 100 까지 합계를 구한 뒤 출력

i = 1;
result = 0
while(i < 101):
    result += i
    i += 1
print(f"1~100 까지의 합: {result}")
```

```python
# 사용자로부터 입력받은 숫자가 4가 아니라면 반복
# isdigit 은 특정 문자 혹은 문자열이 숫자인지 판별하는 함수. 모든 문자가 숫자면 True

i = 0
while(i != 4):
    i = input()
    if i.isdigit():
        i = int(i)
        print(i)
    else:
        print("숫자를 입력하세요")
```

```python
coffee = 10

# break : 반복문을 강제로 빠져나가기

while(1):
    print("커피 주기")
    coffee -=1
    print(f"남은 커피 개수 {coffee}")

    if(coffee == 0):
        print("커피 판매 종료")
        break
```

```python
# 1 ~ 100 사이의 숫자 중에서 3의 배수만을 더한 후 결과값 출력

i = 1
result = 0
while(i<101):
    if(i%3 == 0):
        result += i
    i += 1

print(result)
```

```python
# 1 ~ 100 사이의 숫자 중에서 3의 배수만을 더한 후 결과값 출력 (단 9의 배수는 더하지 않음)

i = 1
result = 0

while(i<101):
    if(i%3 == 0 and i%9 !=0):
        result += i
    i += 1

print(result)
```

```python
# 주차장 프로그램
# 사용자에게 메뉴 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 1
# 자동차 A~
# A 자동차 입차. 주차장 상태 ==> ['A']
# 1 
# 자동차 B
# B 자동차 입차. 주차장 상태 ==> ['A', 'B']

# 2
# 출차 => 뒤에서밖에 못나감
# B 자동차 출차, 주차장 상태 ==> ['A','B']

# 최대 5대까지
# 이후 1 선택시 주차장 만차

space = []
rego = 65
top = 0
while(True):
    x = int(input("[1] 주차 | [2] 출차 | [3] 종료"))
    if(x==1):
        if(len(space) == 5):
            print("만차입니다")
            continue

        # TODO: 입차하기
        space.append(chr(rego))
        print(f"{chr(rego)} 자동차 입차. 주차장 상태 {space}")
        rego += 1

    elif(x==2):
        if(len(space) == 0):
            print("주차장이 비어 있습니다.")
            continue

        # TODO: 출차하기
        if(rego != 65): rego -= 1
        space.pop()
        print(f"{chr(rego)} 자동차 출차. 주차장 상태 {space}")

    elif(x ==3):
        print("프로그램 종료")
        break;

    else:
        print("번호를 확인해 주세요.")
```

#### List Comprehension
- 리스트 안에 for 문을 포함

```python
# 1 ~ 6 숫자를 가지는 리스트 생성

# numbers = [1,2,3,4,5,6]

# 빈 리스트 생성 후 1~6 숫자를 가지는 리스트
# numbers = []
# numbers.append(1)
# numbers.append(2)
# numbers.append(3)
# numbers.append(4)
# numbers.append(5)
# numbers.append(6)

numbers = []

for i in range(1,7):
    numbers.append(i)

print(numbers)
```

```python
# 새로운 리스트 생성 : 리스트 컴프리헨션

[i for i in range(1,7)]
[i*i for i in range(5)]
print([i*2 for i in [1,3,7,9,16]])
print([[i, i+1, i+2] for i in [1,2,3]])

# 1 ~ 10 사이의 숫자 중에서 홀수 리스트 생성
odd_list = []
for i in range(1,11):
    if i % 2 == 1:
        odd_list.append(i)
print(odd_list)

# 리스트 컴프리헨션으로!

odd_list2 = [i for i in range(1,11) if i % 2 == 1]
print(odd_list2)
```

```python
q_list = ["갑", "을", "병", "정", "경"]

# q_list 에서 정을 제외한 요소를 갖는 새로운 리스트 생성
# for 문
new_list = []
for i in q_list:
    if i == "정":
        continue
    new_list.append(i)
print(new_list)

# 리스트 컴프리헨션
new_list1 = [i for i in q_list if i != "정"]
print(new_list1)
```

```python
q1 = ['nice', 'study', 'python', 'anaconda', '1']

# 5글자 이상의 단어만 새로운 리스트로 생성
# len()

new_q1 = []
for i in q1:
    if len(i) >= 5:
        new_q1.append(i)
print(new_q1)

new_q1_1 = [i for i in q1 if len(i)>=5]
print(new_q1_1)


```

```python
q2 = ['A', 'b', 'c', 'D', 'e', 'F', 'G', 'h']

# 소문자만 추출 후 새로운 리스트로 생성

# "X".islower()
# "X".isupper()

new_q2 = [i for i in q2 if i.islower()]
print(new_q2)
```
