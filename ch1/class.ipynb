```python
# 자동차 클래스 설계
# 속성 : 색상, 속도 => 멤버 변수
# 동작 : 전진, 후진, 가속, 감속 => 멤버 메소드

class Car:
    def __init__(self, color, speed):
        # 멤버 변수
        self.color = color
        self.speed = speed

    # 인스턴스(멤버) 메소드(함수)
    def upSpeed(self, value):
        self.speed += value

    def downSpeed(self, value):
        self.speed -= value

    # 0x000001F81B3F5BD0 => 이런 형태를 보고 싶은 문자로 변경
    def __str__(self):
        return f"Car(color = {self.color}, speed = {self.speed})"
```

|`@staticmethod`|없음| 없음(매개변수로 받은 것만) | 클래스와 관련은 있지만 독립적인 계산|

```python
# Account 클래스 작성
# 멤버변수 : 계좌주(owner), 잔액(balance)
# 멤버메서드 : deposit(money) : 잔액 = 잔액 + 입금액
# 멤버메서드 : withdraw(money) : 잔액 = 잔액 - 출금액
# 멤버메소드 : showBalance() : 현재 잔액 출력

class Account:

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, money):
        self.balance += money

    def withdraw(self, money):
        if self.balance < money:
            print("잔액이 부족합니다.")
        else: 
            self.balance -= money

    def showBalance(self):
        print(f"현재 잔액은 {self.balance}입니다.")

acc1 = Account("홍길동", 100000)
acc1.deposit(5000)
acc1.withdraw(15000)
acc1.showBalance()
```

```python
# Order 클래스 작성
# 클래스 변수: total_order
# 멤버 변수: 메뉴(menu), 가격(price), 수량(count)
# 메소드: getTotalPrice() => 가격 x 수량 : 총 주문금액 반환
#             showOrder() : 출력문 작성
#             메뉴 : 아메리카노
#             가격 : 4000
#             수량 : 2개
#             총 주문금액 : 8000원

class Order:

    total_order = 0

    def __init__(self, menu, price, count):
        self.menu = menu
        self.price = price
        self.count = count
        Order.total_order += 1

    def getTotalPrice(self):
        return self.price * self.count

    def showOrder(self):
        print(f"메뉴: {self.menu}")        
        print(f"가격: {self.price}")        
        print(f"수량: {self.count}")
        print(f"총 주문금액: {self.getTotalPrice()}원")

or1 = Order("아메리카노", 4000, 3)        
or1.showOrder()
print()

or2 = Order("카페라떼", 4500, 10)
or2.showOrder()

print(f"전체 주문 수 : {Order.total_order}")

    
```

```python
class Parent:
    def __init__(self):
        self.value = "테스트"
        print("Parent 클래스의 __init__() 호출")

    def test(self):
        print("Parent 클래스의 test() 호출")

class Child(Parent):
    pass

child = Child()     # __init__() 무조건 호출
child.test()
print(child.value)
```

```python
# super() : 상위 클래스

class Child(Parent):
    def __init__(self):
        # 자식 코드에 __init__(self)가 있을경우 아래줄이 반드시 필요
        super().__init__()
        print("Child 클래스의 __init__ 호출")

child = Child()
child.test()
print(child.value)
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name}가 먹이를 먹는다.")

    def sound(self):
        print(f"{self.name}이 소리를 낸다.")

class Dog(Animal):
    def __init__(self, name, kind):
        super().__init__(name)
        self.kind = kind

    def fetch(self):
        print(f"{self.name}이 공을 물어옵니다.")

dog = Dog("초코", "푸들")
dog.eat()
dog.sound()
dog.fetch()
```

```python
# Audio : 맴버변수 : power, volume / 메소드 : 전원조절, 볼륨조절 + tune()
# Tv : 멤버변수 : power, volume, size / 메소드 : 전원조절, 볼륨조절 + watch()

class AudioVisual: 
    def __init__(self, power, volume):
        self.power = power
        self.volume = volume

    def switch(self, on_off):
        self.power = on_off

    def setVolume(self, volume):
        self.volume = volume    

class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self):
        print("La la la..." if self.power else "turn it on")

class TV(AudioVisual):
    def __init__(self, power, volume, size):
        super().__init__(power, volume)
        self.size = size

    def watch(self):
        print("have fun!!" if self.power else "switch on")

tv = TV(False, 12, 50)
# tv 켜기
tv.switch(True)
# tv 보기
tv.watch()

audio = Audio(True, 15)
audio.setVolume(10)
audio.tune()
audio.switch(False)
```

#### 오버라이딩
- 부모 클래스에 있는 메서드를 자식 클래스에서 같은 이름으로 다시 정의해서 동작 내용 변경
- 부모 클라스는 자식의 '최소 사양'
- 자식은 부모보다 '더 많은 것'을 가짐

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f"{self.name}가 소리를 냅니다.")

class Dog(Animal):
    def sound(self):
        print(f"{self.name} : 멍멍!!")
        # return super().sound()


class Cat(Animal):
    def sound(self):
        print(f"{self.name} : 야옹!!")
        # 오버라이딩 시 동일한 부모 메소드를 호출하고 싶다면 super()를 사용한다
        super().sound()

dog = Dog("초코")
cat = Cat("몰리")

dog.sound()
cat.sound()
```

```python
# Employee 부모 클래스 : 
# 멤버 변수 : 이름(name), 급여(pay)
#     메소드 : showInfo() => 이름, 급여 출력

# Manager 자식 클래스
# 멤버 변수 : 보너스(bonus)
# 메소드 : getTotalPay() : 급여 + 보너스 출력

class Employee:
    def __init__(self, name, pay):
        self.name = name
        self.pay = pay

    def showInfo(self):
        print(f"이름: {self.name}, 급여: {self.pay}원")

class Manager(Employee):
    def __init__(self, name, pay, bonus):
        super().__init__(name, pay)
        self.bonus = bonus

    def getTotalPay(self):
        print(f"총 급여 : {self.pay + self.bonus}원")

emp = Employee("성춘향", 300)
emp.showInfo()

emp1 = Manager("홍길동", 10000, 5000)
emp1.showInfo()
emp1.getTotalPay()
```

```python
# Shape 부모 클래스
# 메소드 : area() => "도형의 넓이" 출력

# Rectangle 자식 클래스
# 멤버 변수 : width, height
# 메서드 : area() => 직사각형의 넓이 출력

# Circle 자식 클래스
# 멤버 변수 : radius
# 메서드 : area() => 원의 넓이 출력

class Shape:

    def area(self):
        print(f"도형의 넓이")

class Rectangle(Shape):

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        print(f"직사각형의 넓이 : {self.width*self.height}")

class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print(f"원의 넓이 : {self.radius**2*3.14}")

rec1 = Rectangle(10, 20)
rec1.area()

cir1 = Circle(10)
cir1.area()
        
```
