````python
# 일반 매개변수는 앞 쪽으로 선언
# +, * 실행

def add_mul(choice, a, *args):
    if choice == "add":
        result = 0
        for i in args:
            result = result + i
    elif choice == "mul":
        result = 1
        for i in args:
            result = result * i
            
    return result
````

````python
# 키워드 매개변수 (key = value)
# **변수명

# 딕셔너리 == 키워드 매개변수
def print_kwargs(**kwargs):
    print(kwargs)

    # key값만 출력
    for k in kwargs.keys():
        print(k)

    # value값만 출력
    for v in kwargs.values():
        print(v)
    
    # key, value 같이 출력
    for k,v in kwargs.items():
        print(k,v)
````

````python
print_kwargs(name="kim", age=15)
print_kwargs(name="park", age=23, addr = "Seoul")
print_kwargs(name="hong", age=18, addr = "kyeonggi", tel = "010-1234-5678")
````

````python
# 전역변수 a값 변경하기
# 1) 리턴 이용(추천!)

a = 1
def vartest(a):
    # 매개변수 a : 지역변수
    a = a + 1
    return a

a = vartest(a)
print(a)
````

````python
# 2) global 키워드 사용!
# 함수 안에서 전역변수 접근 안됨! == 파이썬만의 특징

a = 1
def vartest():
    global a
    a = a + 1

vartest()
print(a)
````

````python
# 리스트의 평균 구하는 함수
# get_average()
# 요소의 개수는 정해지지 않음
# 리턴하지 말고 평균 출력


def get_average(*args):
    result = 0
    for i in args:
        result = result + i
    print(result / len(args))

get_average(10,11,12)
get_average(10,11,12,15,25)
get_average(10,11,12,15,25,36)
````

````python
def basic_calculator(num1, num2, op):
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "/":
        return num1 / num2
    elif op == "//":
        return num1 // num2
    
print(basic_calculator(10, 20, "+"))
print(basic_calculator(10, 20, "-"))
print(basic_calculator(10, 20, "*"))
print(basic_calculator(10, 20, "/"))   
print(basic_calculator(10, 20, "//"))   
````

````python
# 사칙연산 함수
# 결과 리턴!

def calculator(*args):
    # if ~ elif

    choice = args[-1]
    numbers = args[:-1]
    result = numbers[0]
    
    if choice == "+":
        for i in numbers[1:]:
            result += i
    elif choice == "-":
        for i in numbers[1:]:
            result -= i
    elif choice == "*":
        for i in numbers[1:]:
            result *= i
    elif choice == "/":
        for i in numbers[1:]:
            result /= i
    return result

        

print(calculator(10, 20, "+"))
print(calculator(10, 20, "-"))
print(calculator(10, 20, "*"))
print(calculator(10, 20, "/"))
````
