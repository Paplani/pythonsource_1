```python
num = input("숫자 입력")

# type(num)
# 사용자에게 숫자를 입력받은 후 +10 한 결과 출력
# try ~ except

if num.isdigit():
    num = int(num) + 10
    print(num)
else:
    print("숫자를 입력하세요")
```
