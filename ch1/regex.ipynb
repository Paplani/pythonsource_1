#### 정규표현식(정규식)
- 복잡한 문자열을 처리할 때 사용하는 방법
- 파이썬만의 고유한 문법이 아니라 문자열을 처리하는 모든 곳에서 사용하는 일종의 형식 언어

```python
# 주민등록번호 => 뒷자리를 모두 * 문자로 변경

# 파이썬에서 정규식 지원하는 모듈

import re

dir(re)

# re.compile() : 정규식을 작성
pattern = re.compile("ab*")

# 패턴을 특정 문자열이 포함하고 있는가?       
print(pattern.match("abc"))
```

| 메서드 | 목적 |
|-----|-----|
|`match` | 문자열의 처음부터 정규식과 매치되는지 조사|
|`search` | 문자열 전체를 검색하여 정규식과 매치되는지 조사|
|`findall` | 정규식과 매치되는 모든 문자열(substring)을 리스트로 리턴|
|`finditer` | 정규식과 매치되는 모든 문자열(substring)을 이터레이터 객체로 리턴|

```python
# [문자] : [] 안에 들어있는 문자들과 매치하느냐??

p = re.compile("[abc]")

print(p.match("a"))
print(p.match("before"))
print(p.match("dude"))
```

```python
m = p.match("a")

print(f"패턴 시작 : {m.start()}")
print(f"패턴 끝 : {m.end()}")
print(f"패턴과 일치하는 문자열 반환 : {m.group()}")
print(f"일치의 시작, 끝 반환 : {m.span()}")

if m:
    print(f"Match found {m.group()}")
else:
    print("No Match")
```

```python
m = p.search("a")

print(m)
```

```python
print(p.search("3 before"))
# match는 처음부터 일치해야함
print(p.match("3 before"))
```

```python
p = re.compile("[abc]")
m = p.findall("Happy forever")
print(m)

# finditer를 통해 받으면 객체로 받을수 있음. for문 가능
m = p.finditer("Happy forever")
print(m)

for r in m:
    print(r)
```

```python
# .문자 : \n 을 제외한 모든 문자

p = re.compile("a.b")

print(p.match("a\nb"))
print(p.match("abb"))
print(p.match("a0b"))
print(p.match("aab"))
print(p.match("a b"))
```

```python
# . : 마침표로서 사용하고 싶을 때
# \. or [.] : 이래야 마침표로 인식함
# \t, \r ... : 이스케이프 문자로 인식할수 있기 때문에 앞에 r을 써줘야함

p = re.compile(r"a\.b")
# p = re.compile("a\.b") => 오류뜸

print(p.match("a\nb"))
print(p.match("abb"))
print(p.match("a0b"))
print(p.match("aab"))
print(p.match("a.b"))
```

```python
# * : 0 ~ 무한대
# + : 1 ~ 무한대
# ? : 0 ~ 1

# ca
p = re.compile("ca*t")
print(p.match("ct"))
print(p.match("cat"))
print(p.match("caaaaaaaaaaat"))
```

```python
p = re.compile("ca+t")
print(p.match("ct"))
print(p.match("cat"))
print(p.match("caaaaaaaaaaat"))
```

```python
p = re.compile("ca?t")
print(p.match("ct"))
print(p.match("cat"))
print(p.match("caaaaaaaaaaat"))
```

```python
# a가 2개
p = re.compile("ca{2}t")
print(p.match("ct"))
print(p.match("cat"))
print(p.match("caat"))
print(p.match("caaaaaaaaaaat"))
```

```python
p = re.compile("ca{2,5}t")
print(p.match("ct"))
print(p.match("cat"))
print(p.match("caat"))
print(p.match("caaaaaaaaaaat"))
```

```python
# a ~ z 모든 소문자
p = re.compile("[a-z]")

print(p.search("kkk1234"))
print(p.search("k1234ab14c"))

p = re.compile("[a-z]+")
print(p.search("kkk1234"))
print(p.search("k1234ab14c"))

print()

# [^ ] : 대괄호 안에 있는 문자를 제외한
p = re.compile("[^a-z]+")
print(p.search("kkk1234"))
print(p.search("k1234ab14c"))

# 0, 대소문자, 숫자를 제외한 문자
p = re.compile("[^a-zA-Z0-9]+")
print(p.search("kkk1234"))
print(p.search("k1234ab14c"))
print(p.search(" python"))
```
