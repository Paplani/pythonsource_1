```python
# 오류 강제 발생 : raise 오류명

class Bird:
    def fly(self):
        raise NotImplementedError

class Eagle(Bird):
    def fly(self):
        print("Very fast")

eagle = Eagle()
eagle.fly()
```

```python
# 예외 클래스 생성 가능
class MyError(Exception):
    def __str__(self):
        return "허용되지 않는 변명입니다."

def say_nick(nick):
    if nick == "바보":
        raise MyError()
    print(nick)

try:
    say_nick("천사")
    say_nick("바보")
except MyError as e:
    print(e)

```

```python
# 문제 1. 랜덤 비밀번호 생성하기

# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램 작성
# choice()
chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz!@#$"
# 비밀번호 길이는 4이상만 허용
# 숫자가 아닌 값을 입력하면 예외 처리

import random

try:
    length = int(input("생성할 비밀번호의 길이를 입력해주세요."))
    chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz!@#$%^&*"
    password = ""
    if length >= 4:
        password = random.choices(chars, k=length)
        result = "".join(password)
        print(result)
    else:
        print("비밀번호의 길이는 4 이상이여야합니다.")
    
except ValueError:
    print("숫자를 입력해주세요.")   
```

```python
# 솔루션 해답

import random

while True:
    try:
        pwd_num = int(input("비밀번호 길이 : "))

        if pwd_num < 4:
            print("비밀번호는 4자리 이상이여야 합니다")
            continue
        break

    except ValueError:
        print("숫자만 입력하세요")

chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz!@#$%^&*"
password = []
for i in range(pwd_num):
    password.append(random.choice(chars))

# 리스트 컴프리헨션
# password = [random.choice(chars) for i in range(pwd_num)]
print(password)
print(f"랜덤 비밀번호 : {"".join(password)}")
# 리스트 => 문자열로 변경 : .join() 사용!
```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력형식 2000-08-25
# input() 으로 입력받은건 문자열로 들어옴. .strptime()사용!

from time import time, strptime
from datetime import datetime

inputBirth = input("생년월일을 입력해주세요")

try:
    birthday = datetime.strptime(inputBirth, "%Y-%m-%d")
    today = datetime.now()
    age = today.year - birthday.year
    if (today.month, today.day) < (birthday.month, birthday.day):
        age -= 1
    print(age)
except ValueError:
    print("생년월일을 알맞게 입력해주세요")
```

```python
# 학생 정보를 json으로 저장하기
# 학생정보를 입력받아 json 형식으로 파일 저장
# 이름, 나이, 전공
# 읽어서 출력

import json

inputData = input("학생정보를 입력해주세요 / 이름, 나이, 전공 순으로 입력해주세요.")
data = inputData.split(",")
dict1 = {"name":data[0], "age": data[1], "major":data[2]}

with open("./data/stu_info.json", "w", encoding = "utf-8") as f:
    json.dump(dict1, f, ensure_ascii=False)

with open("./data/stu_info.json", "r", encoding = "utf-8") as f:
    readData = json.load(f)
    print(f"이름: {readData['name']}")
    print(f"나이: {readData['age']}")
    print(f"전공: {readData['major']}")
```

```python
# 다음과 같은 csv 파일이 있다.
# 이름, 국어, 영어, 수학
# 홍길동, 90, 80, 70
# 김철수, 16, 40, 65
# 이영희, 90, 53, 39
# 성춘향, 95, 50, 73

# csv 파일을 읽어 학생별 평균 점수 출력
# 홍길동 : 80.0

# csv.DicReader() 사용, 평균은 둘째자리까지

import csv

with open("./data/sample5.csv", "w", encoding="utf-8-sig", newline="") as f:
    headers = ['이름', '국어', '영어', '수학']
    writer = csv.DictWriter(f, fieldnames=headers)

    writer.writeheader()
    writer.writerow({"이름":"홍길동", "국어":90, "영어":82, "수학":80})
    writer.writerow({"이름":"김철수", "국어":97, "영어":69, "수학":96})
    writer.writerow({"이름":"이영희", "국어":95, "영어":93, "수학":83})
    writer.writerow({"이름":"성춘향", "국어":86, "영어":85, "수학":80})

with open("./data/sample5.csv", "r", encoding="utf-8-sig") as f:
    reader = csv.DictReader(f)
    # next(reader) : DictReader에서는 쓸 필요가 없음
    for i in reader:
        avg = sum([int(i["국어"]), int(i["영어"]), int(i["수학"])])/3
        # avg = total/3
        result = round(avg, 2)
        print(f"{i["이름"]} : {result:.2f}")
```
