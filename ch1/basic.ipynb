#### 문자열 포맷팅 (formatting)
- 문자열 안에 어떤 값을 삽입하는 방법

| 코드 | 설명 |
| --- | --- |
| `%s` | 문자열(string) |
| `%d` | 정수(integer) |
| `%c` | 문자 1개(character) |
| `%f` | 부동 소수(floating-point) |
| `%o` | 8진수 |
| `%x`| 16진수 |
| `%%` | % 문자 자체 |

- format() 함수
- f 문자열 포매팅 (python 3.6버전부터 지원)

```python
a = "I eat %d apples" %3
a
```

```python
# %s => 어떤 값이든 대입 가능함

a = "I eat %s apples" %"five"
a
```

```python
a = "Error is %d%%" %98
a
```

```python
a = "I ate %d apples, so I was sick for %s days" %(3, "four")
a
```

```python
# hi 문자열 출력시 10개인 공간 사용
# 오른쪽 정렬

print("%10s" % "hi")

# 총 10개인 칸을 쓰는데 왼쪽 정렬! -붙이면 됨
print("%-10sjane" % "hi")

# 숫자 0은 길이에 상관하지 않음
# 소수점 네번째 자리까지만
print("%0.4f" % 3.42134234)

# 전체 길이는 10개인 문자열 공간에서 오른쪽으로 정렬
print("%10.4f" % 3.42134234)
```

```python
a = "I eat {} apples" .format(3)
a
```

```python
a = "I ate {0} apples, so I was sick for {1} days" .format(3, "four")
a
```

```python
name = "홍길동"
age = 30

# 첫 번째 방법
print("나의 이름은 %s 이고, 나이는 %d 입니다" %(name, age))

# 두번째 방법
print("나의 이름은 {} 이고, 나이는 {} 입니다" .format(name, age))

# 세번째 방법
print(f"나의 이름은 {name} 이고, 나이는 {age+1} 입니다" .format(name, age))
```

```python
# 공백제거

print('hi' == '  hi  ')

a = '  hi  '

#왼쪽 공백 제거
print(a.lstrip())

# 오른쪽 공백 제거
print(a.rstrip())

# 양쪽 공백 제거
print(a.strip())

print('hi' == a.strip())

print(a)
```
