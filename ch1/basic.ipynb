#### 문자열 포맷팅 (formatting)
- 문자열 안에 어떤 값을 삽입하는 방법

| 코드 | 설명 |
| --- | --- |
| `%s` | 문자열(string) |
| `%d` | 정수(integer) |
| `%c` | 문자 1개(character) |
| `%f` | 부동 소수(floating-point) |
| `%o` | 8진수 |
| `%x`| 16진수 |
| `%%` | % 문자 자체 |

- format() 함수
- f 문자열 포매팅 (python 3.6버전부터 지원)

```python
a = "I eat %d apples" %3
a
```

```python
# %s => 어떤 값이든 대입 가능함

a = "I eat %s apples" %"five"
a
```

```python
a = "Error is %d%%" %98
a
```

```python
a = "I ate %d apples, so I was sick for %s days" %(3, "four")
a
```

```python
# hi 문자열 출력시 10개인 공간 사용
# 오른쪽 정렬

print("%10s" % "hi")

# 총 10개인 칸을 쓰는데 왼쪽 정렬! -붙이면 됨
print("%-10sjane" % "hi")

# 숫자 0은 길이에 상관하지 않음
# 소수점 네번째 자리까지만
print("%0.4f" % 3.42134234)

# 전체 길이는 10개인 문자열 공간에서 오른쪽으로 정렬
print("%10.4f" % 3.42134234)
```

```python
a = "I eat {} apples" .format(3)
a
```

```python
a = "I ate {0} apples, so I was sick for {1} days" .format(3, "four")
a
```

```python
name = "홍길동"
age = 30

# 첫 번째 방법
print("나의 이름은 %s 이고, 나이는 %d 입니다" %(name, age))

# 두번째 방법
print("나의 이름은 {} 이고, 나이는 {} 입니다" .format(name, age))

# 세번째 방법
print(f"나의 이름은 {name} 이고, 나이는 {age+1} 입니다" .format(name, age))
```

```python
a = "java"
b = [1,2,3,4]
c = (1,2,3)
d = {"a":1}

# e = a 로 하는것과 e = "java"로 하는거의 차이점은?
# 문자열은 불변객체임. 문자열은 생성되면 그 내용을 바꿀 수 없고 값을 바꾸려고 하면 메모리에 새로운 문자열을 만듬
e = a
print(a, e)
a = "python"
print(a,e)
```

```python
# 리스트 복사할때 주의할 점
# 리스트는 수정 가능한(Mutable)객체기 때문에 e = a 와 e = [값] 일때 값이 같이 바뀔수 있다.
# 추가로 딕셔너리도 수정가능한 자료형임!

list1 = [1,2,3]
# 이건 복사가 아니라 같은 곳을 바라보게 하는 것
list2 = list1
list1.append(4)
print(list1, list2)
```

```python
# 리스트 복사
# 방법 1

list1 = [1,2,3]
list2 = list1[:]
list1.append(4)
print(list1, list2)
```

```python
# 방법 2

list1 = [1,2,3]
list4 = list1.copy()
list1.append(4)
print(list1, list4)
```

```python
# 방법 3

from copy import copy

list1 = [1,2,3]
list3 = copy(list1)
list1.append(4)
print(list1, list3)
```
