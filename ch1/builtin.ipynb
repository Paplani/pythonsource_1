```python
# abs(x) : 절대값 리턴

print(abs(3))
print(abs(-3))
print(abs(3.14))
```

```python
# chr() <=> ord()
# chr() : 유니코드 값을 넣으면 그 코드에 해당하는 문자를 리턴
# ord() : 문자를 넣으면 그 문자에 해당하는 유니코드를 리턴

print(chr(97))
print(chr(44023))

print(ord('a'))
print(ord('가'))
```

```python
# divmod(a,b) : 몫과 나머지를 튜플로 반환

quot, rem = divmod(7,3)
print(quot)
print(rem)
```

```python
# enumerate() : 리스트, 튜플, 문자열을 입력으로 받아 인덱스 값을 포함하는 enumerate 객체 반환

for i, name in enumerate(['body', 'foo', 'bar'], start = 1):
    print(i, name)

for i, name in enumerate('bodybar', start = 1):
    print(i, name)
```

```python
print(max([1,2,3]))
print(max('python'))
print(min([1,2,3]))
print(min('python'))
```

```python
# round(숫자, [ndigits]) : 반올림

print(round(4,6))
print(round(4.24241, 3))
```

```python
# sort() : 원본이 변한다는 단점이 있음
list1 = [1,2,6,3,4,8]
list2 = [1,2,6,3,4,8,9]
# 정렬
list1.sort()
print(list1)

# sorted() : 입력 데이터를 정렬후 결과를 새로운 리스트로 리턴
print(sorted(list2))
print(list2)
```

```python
# zip() : 동일한 개수로 이루어진 데이터들을 묶어서 반환해주는 함수

print(list(zip([1,2,3], [4,5,6])))
print(list(zip([1,2,3], [4,5,6], [7,8,9])))
print(list(zip("abc", "def")))
```

```python
# 학생 5명의 점수를 입력받아 다음 정보를 출력하는 프로그램을 작성하시오

# 출력내용
# 최고점수, 최저점수, 평균점수(소수 둘째 자리), 점수를 오름차순으로 출력, 등수와 함께 출력
# max, min, sum, len, sorted, enumerate

scores = list(map(int, input("5명의 점수를 입력해주세요.").split(",")))
print(scores)

topScore = max(scores)
lowScore = min(scores)
avgScore = sum(scores)/5
acsScore = sorted(scores)
dcsScore = sorted(scores, reverse = True)

print(f"최고점수 : {topScore}")
print(f"최저점수 : {lowScore}")
print(f"평균점수 : {avgScore}")
print(acsScore)
print(dcsScore)

# 등수
for idx, jumsu in enumerate(dcsScore, 1):
    print(f"{idx}등 : {jumsu}")

```

```python
# 문자열을 입력받아 다음 정보를 출력
# 출력내용 : 입력한 문자, 각 문자의 유니코드 값, 가장 큰 문자, 가장 작은 문자

# ord, chr, max, min, list
# 문자열 입력 : Python
# P -> 80
# y -> 121
#
# 가장 큰 문자 : y
# 가장 작은 문자 : P

word = input("글자를 입력해주세요")

print(list1)

for c in word:
    uniLetter = ord(c)
    print(f"{c} -> {uniLetter}")

print(f"가장 큰 문자 : {max(word)}")
print(f"가장 작은 문자 : {min(word)}")
```

```python
x = [3,7,1,-5,9]
y = [2,-4,8,10,-1]

# 절댓값을 더해라!
# zip, abs, enumerate
# 출력
# (3,2) => 5

# zip을 써서 각 순서끼리 먼저 튜플의 형태로 만듬.
for i, (a,b) in enumerate(zip(x,y)):
    result = abs(a) + abs(b)
    print(f"({a},{b}) => {result}") 
```

```python
# 사용자로부터 숫자 10개 입력받기
# 출력
# 입력받은 숫자 리스트, 중복 제거, 오름차순, 튜플변환, 집합 크기
# 원본 [~~~~]
# 중복 제거 {~~~}
# 정렬
# 튜플
# 집합 크기 : len()

nums = list(map(int, input("숫자 10개를 입력해주세요. (숫자,숫자,숫자)형태로 해주세요.").split(",")))

ascNums = sorted(nums)
dcsNums = sorted(nums, reverse = True)

print(nums)
print(list(set(nums)))
print(ascNums)
print(dcsNums)
print(tuple(nums))
print(len(nums))

```
