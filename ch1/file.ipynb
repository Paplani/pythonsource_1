#### file 읽고 쓰기

| 파일 열기 모드 | 설명 |
|------------|-------|
|`r`| **읽기(Read) 모드** : 파일의 내용을 읽기만 할때 사용. 파일이 존재하지 않는 경우 오류 발생. |
|`w`| **쓰기(Write) 모드** : 파일의 내용을 새로 쓸때 사용. 기존 파일이 존재하는 경우 내용 삭제 후 새로 작성. 없는 경우 새로 생성. |
|`a`| **추가(Append) 모드** : 기존 파일의 마지막에 내용을 추가할 때 사용. 파일이 없는 경우 새로 생성. |

```python
# score.txt 내용을 읽은 후 평균을 구한 후 출력

scores = []
with open("./data/score.txt", "r") as f:
    for num in f:
        scores.append(int(num))

print(scores)
print(f"평균 : {sum(scores) / len(scores)}")
```

```python
# sample.txt 파일을 읽어서 총합과 평균을 구한 후 총합과 평균을 result.txt 파일에 작성
# 아래의 방식은 total 이라는 리스트에 담아둬서 메모리를 차치함

total = []
with open("./data/sample.txt", "r") as f:
    for num in f:
        total.append(int(num))
    total_score = sum(total)
    average = total_score / len(total)

with open("./data/result.txt", "w", encoding = "utf-8") as f:
    f.write(f"총합: {total_score}, 평균:{average:.2f}\n")
```

```python
# fruits.txt 읽어서 각각의 과일 단어가 몇번 나왔는지
# apple 3
# banana 2

# apple banana orange
# 단어 분리 : .split()
# split()을 사용하면 리스트로 나옴    ['apple', 'banana', 'orange']

# dic 추가
# dic1['apple'] += 1

with open("./data/fruits.txt", "r") as f:
    line = f.read().split()
    print(line)

    dic1 = {}
    for word in line:
        if word in dic1:
            dic1[word] += 1       
        else:
            dic1[word] = 1
    
    print(dic1)

for fruit, count in dic1.items():
    print(f"{fruit} : {count}")
    
```

```python
# 1000 명의 이름, 키, 몸무게 info.txt. 읽은 후
# BMI 지수 계산
# bmi = 체중 / 키의 제곱
# bmi가 18.5 미만일 때 저체중, 22.9 미만일떄 정상, 24.9 과체중, 25이상 비만
# 출력문 형태
# 이름 : 하타
# 몸무게 : 95
# 키 : 180
# BMI : 20.0
# 결과 : 정상체중

person_dict = {}
with open("./data/info.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, weight, height = line.strip().split(", ")
        weight = int(weight)
        height = int(height)
        bmi = weight / (height/100)**2

        if bmi < 18.5 :
            result = "저체중"
        elif 18.5<= bmi < 22.9:
            result = "정상"
        elif 22.9 <= bmi < 24.9:
            result = "과체중"
        else:
            result = "비만"

        # print("\n".join(["이름: {}", "몸무게: {}", "키: {}", "BMI: {:.1f}", "결과: {}"]).format(name, weight, height, bmi, result))

        # 아래는 딕셔너리로 관리 : 이름을 알면 정보 다 튀어나오게!
        person_dict[name] = {
            "키": height,
            "몸무게": weight,
            "BMI" : round(bmi, 2),
            "결과" : result
        }
     
```

```python
# origin.txt 읽어서 암호화 시킨 후 encr.txt 저장 
# 1. 암호화 선택 : origin.txt 읽어서 변수에 담기  => 그걸 chr(ord("안")+100)
# 2. 복호화 선택 : 암호화된 파일을 읽은 후 원래대로 출력

while True:
    no = int(input("1.암호화 | 2.복호화 | 3.종료 중 원하는 메뉴의 번호를 입력하세요"))

    if no == 1:
        with open("./data/origin.txt", "r", encoding="utf-8") as f_in:
            contents = f_in.read()

            with open("./data/encr.txt", "w", encoding="utf-8") as f_out:
                for c in contents:
                    f_out.write(chr(ord(c)+100))
    elif no == 2:
        with open("./data/encr.txt", "r", encoding = "utf-8") as f:
            contents = f.read()
            for c in range(len(contents)):
                print(chr(ord(contents[c])-100), end="")
    else:
        break
```
